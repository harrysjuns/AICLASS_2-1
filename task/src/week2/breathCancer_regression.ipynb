# BreathCancer Classification

Linear regression practice notebook for the breath cancer dataset.
> - 회귀: 연속되는 값 자체를 예측
> - label 을 제거한 데이터로 cancer(암)의 면적(진행정도) 정도를 예측하는 실습

### 도구 설정 및 데이터 불러오기
- 선형 회귀 모델을 불러옴 (예측하려는 값이 연속적이기 때문)
- 회귀 모델의 성능을 평가하기 위한 지표인 MSE(mean_squared_error)를 계산하는 함수를 불러옴
- MSE는 실제값과 예측값의 차이의 제곱의 평균을 의미함

```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
path = '/Users/ganghyejun/Desktop/100_College/2-1/인공지능개론/AICLASS_2-1/task/src/data/breast_cancer.csv'
df = pd.read_csv(path) 
df2=df.drop('label',axis=1) # 당뇨 레이블은 제거 (연속적이지 않은 값인 Outcome 열 제거)
df2
```

### 데이터 정리
- worst area 열을 예측 목표 y로 지정
- worst area 열을 제거한 나머지 데이터를 X로 지정
- train_test_split을 이용해 데이터를 훈련용과 테스트용으로 나눔
- test_size=0.25 : 전체 데이터의 25%를 테스트용으로 사용
- random_state=0 : 데이터를 무작위로 나누되, 실행할 때마다 같은 방식으로 나뉘도록 고정하는 값

```python
y=df2['worst area']
y.value_counts()

X=df2.drop('worst area',axis=1)
X.head()

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25,random_state=0)
```

### 실행
1. 훈련
- LinearRegression 클래스의 객체를 생성하여 model에 저장
- model 객체의 fit 메소드를 이용해 훈련용 데이터 X_train과 y_train으로 모델을 학습시킴

2. 테스트
- model 객체의 predict 메소드를 이용해 테스트용 데이터 X_test에 대한 예측값을 ly_preds에 저장

3. 성능 평가
- MSE를 이용해 실제값과 예측값을 비교하여 성능을 평가하고 출력
- mse_np는 numpy를 이용한 MSE 구현
- mse는 Python 문법을 이용한 MSE 구현

```python
model = LinearRegression()
model.fit(X_train, y_train)

ly_preds = model.predict(X_test)

print('평균제곱근오차', mean_squared_error(ly_preds, y_test))

def mse_np(actual, predicted):
    return np.mean((np.array(actual) - np.array(predicted)) ** 2)

print('평균제곱근오차', mse_np(ly_preds, y_test))

def mse(actual, predicted): 
    sum_square_error = sum((a - p) ** 2 for a, p in zip(actual, predicted))
    mean_square_error = sum_square_error / len(actual)
    return mean_square_error

print('평균제곱근오차', mse(ly_preds, y_test))
```
