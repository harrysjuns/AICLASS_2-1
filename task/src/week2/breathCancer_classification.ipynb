### 도구 설정 및 데이터 불러오기
1. 데이터 분석을 위해 pandas 를 불러오며 pd 로 줄임(사용편의)
2. 수치 계산을 위해 numpy 를 불러오며 np 로 줄임(사용편의)
3. 로지스틱 회귀 모델을 불러옴 (diabetes 데이터의 outcome이 0과 1로 이루어져 있기 때문에 이진분류에 적합한 모델인 로지스틱 회귀 모델을 사용)
4. 분류 모델의 성능을 평가하기위해 accuracy_score 를 불러옴 (전체적으로 몇개 맞췄는지 평가하는 지표)
5. 분류 모델의 성능을 평가하기위해 confusion_matrix 를 불러옴 (어떤 종류의 실수가 있었는지 평가하는 지표)
6. 데이터를 훈련용과 테스트용으로 나누는 함수 train_test_split 을 불러옴 (train 데이터: 모델 공부시키는 용도 / test 데이터: 공부 안 시킨 새 문제로 성능 확인하는 용도)

```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
path = '/Users/ganghyejun/Desktop/100_College/2-1/인공지능개론/AICLASS_2-1/task/src/data/breast_cancer.csv' #데이터 파일 경로
df = pd.read_csv(path) #pandas 라이브러리를 이용해 csv 파일을 읽어와 df에 저장
df
```

### 데이터 정제
- label 열을 y로 저장 (label 열이 우리가 예측하려는 타겟값이기 때문)
- label 열의 각 값이 몇개씩 있는지 세어서 보여줌 (0과 1이 각각 몇개씩 있는지 확인)

- label 열을 제거한 값을 X로 저장 (,,)

- train_test_split 함수를 이용해 데이터를 훈련용과 테스트용으로 나눔
- test_size=0.5 : 전체 데이터의 50%를 테스트용으로 사용
- random_state=0 : 데이터를 무작위로 나누되, 실행할 때마다 같은 방식으로 나뉘도록 고정하는 값

```python
y=df['label']
y.value_counts()

X=df.drop('label',axis=1)
X.head()

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.5,random_state=0)
```

### 실행
1. 훈련
- LogisticRegression 클래스의 객체를 생성하여 clf_lr에 저장 (훈련하고 테스트를 진행할 모델을 만듬)
- clf_lr 객체의 fit 메소드를 이용해 훈련용 데이터 X_train과 y_train을 이용해 모델을 학습시킴

2. 테스트
- clf_lr 객체의 predict 메소드를 이용해 테스트용 데이터 X_test에 대한 예측값을 pred_lr에 저장

3. 성능 평가
- accuracy_score 와 confusion_matrix 를 사용해 평가 후 지표를 출력

```python
clf_lr = LogisticRegression(random_state=0)
clf_lr.fit(X_train, y_train)

pred_lr = clf_lr.predict(X_test)

print ("\n--- Logistic Regression Classifier ---")
print (accuracy_score(y_test, pred_lr))
print (confusion_matrix(y_test, pred_lr))
```
