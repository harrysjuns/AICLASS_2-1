# Abalon Regression

abalone лҚ°мқҙн„°м…Ӣмқ„ мқҙмҡ©н•ҙ м—°мҶҚнҳ• к°’мқ„ мҳҲмёЎн•ҳлҠ” нҡҢк·Җ л¬ём ңлҘј мҲҳн–үн•ҳмҳҖлӢӨ.
XлҠ” мһ…л Ҙ нҠ№м„ұмңјлЎң, yлҠ” мҳҲмёЎн•  мҲҳм№ҳ лҚ°мқҙн„°лЎң мӮ¬мҡ©н•ҳмҳҖлӢӨ.

- мқҙлІҲ л”Ҙлҹ¬лӢқ нҡҢк·Җ н…ҢмҠӨнҠёмқҳ нҸүк· м ңкіұмҳӨм°Ё(MSE)лҠ” м•Ҫ 4.4 лЎң мёЎм •лҗҳм—ҲлӢӨ.

```python
import pandas as pd
import numpy as np
import tensorflow as tf

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

from tensorflow.keras import layers, models
```

```python
path = '/Users/ganghyejun/Desktop/100_College/2-1/мқёкіөм§ҖлҠҘк°ңлЎ /AICLASS_2-1/task/src/data/abalone.csv'
df = pd.read_csv(path, index_col=0)
df
```

```python
y = df['Rings']
X = df.drop('Rings', axis=1)
```

### 1. лІ”мЈјнҳ• ліҖмҲҳмқё 'Sex'лҘј мӣҗ-н•« мқёмҪ”л”©н•ҳм—¬ мҲ«мһҗнҳ• лҚ°мқҙн„°лЎң ліҖнҷҳ
- abalone лҚ°мқҙн„°мқҳ Sex м»¬лҹјмқҖ M, F, IмІҳлҹј л¬ёмһҗм—ҙлЎң мқҙлЈЁм–ҙм§„ лІ”мЈјнҳ• м •лӢө лҚ°мқҙн„°мқҙлҜҖлЎң, л”Ҙлҹ¬лӢқ лӘЁлҚём—җ к·ёлҢҖлЎң л„Јм–ҙ н•ҷмҠөн•  мҲҳ м—ҶлӢӨ.
- мқҙлҹ° л¬ём ңлҘј н•ҙкІ°н•ҳкё° мң„н•ҙ л¬ёмһҗм—ҙ нҒҙлһҳмҠӨлҘј мҲ«мһҗ нҳ•нғңлЎң л°”кҫёлҠ” мӣҗ-н•« мқёмҪ”л”©мқ„ мӮ¬мҡ©н•ңлӢӨ.
- мӣҗ-н•« мқёмҪ”л”©мқҖ M, F, I к°Ғк°Ғмқ„ лі„лҸ„мқҳ м»¬лҹјмңјлЎң лӮҳлҲ„кі , н•ҙлӢ№ нҒҙлһҳмҠӨл§Ң 1, лӮҳлЁём§ҖлҠ” 0мңјлЎң н‘ңмӢңн•ҳлҠ” л°©мӢқмқҙлӢӨ.
- мқҙ л°©мӢқмқ„ мӮ¬мҡ©н•ҳл©ҙ M=0, F=1, I=2мІҳлҹј нҒҙлһҳмҠӨ мӮ¬мқҙм—җ мӢӨм ңлЎң м—ҶлҠ” к°Җм§ң нҒ¬кё° кҙҖкі„к°Җ мғқкё°лҠ” кІғмқ„ л§үмқ„ мҲҳ мһҲм–ҙ, abalone м„ұлі„ л¶„лҘҳм—җм„ң к°Ғ нҒҙлһҳмҠӨлҘј лҸ…лҰҪм ҒмңјлЎң н•ҷмҠөн•ҳлҠ” лҚ° лҸ„мӣҖмқҙ лҗңлӢӨ.

```python
X = pd.get_dummies(X, columns=['Sex'])
```

```python
# 2. XлҘј numpyлЎң ліҖнҷҳ + мҠӨмјҖмқјл§Ғ
scaler = StandardScaler()
X = scaler.fit_transform(X)

# 3. train/test л¶„лҰ¬
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=0
)
```

```python
# 4. л”Ҙлҹ¬лӢқ нҡҢк·Җ лӘЁлҚё
model = models.Sequential([
    layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    layers.Dense(32, activation='relu'),
    layers.Dense(1)
])

model.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mse']
)

history = model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=50,
    batch_size=16,
    verbose=1
)
```

```python
# 5. мҳҲмёЎ л°Ҹ нҸүк°Җ
y_pred = model.predict(X_test)

print(mean_squared_error(y_test, y_pred))
```
