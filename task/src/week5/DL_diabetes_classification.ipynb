diabetes лҚ°мқҙн„°м…Ӣмқ„ мқҙмҡ©н•ҙ лӢ№лҮЁлі‘ м—¬л¶ҖлҘј л¶„лҘҳн•ҳлҠ” л”Ҙлҹ¬лӢқ л¶„лҘҳ л¬ём ңлҘј мҲҳн–үн•ҳмҳҖлӢӨ.
мһ…л Ҙ лҚ°мқҙн„°(X)мҷҖ м •лӢө лҚ°мқҙн„°(y)лҘј лӮҳлҲҲ л’Ө, yлҘј мӣҗ-н•« мқёмҪ”л”©н•ҳм—¬ л”Ҙлҹ¬лӢқ лӘЁлҚёлЎң н•ҷмҠөн•ҳмҳҖлӢӨ.

- мқҙлІҲ л”Ҙлҹ¬лӢқ л¶„лҘҳ н…ҢмҠӨнҠёмқҳ м •нҷ•лҸ„лҠ” м•Ҫ 0.78лЎң мёЎм •лҗҳм—ҲлӢӨ.
- confusion matrixлҘј ліҙл©ҙ л№„лӢ№лҮЁ нҒҙлһҳмҠӨ(0)лҠ” л№„көҗм Ғ мһҳ л¶„лҘҳлҗҳм—Ҳм§Җл§Ң, лӢ№лҮЁ нҒҙлһҳмҠӨ(1)м—җм„ңлҠ” мқјл¶Җ мҳӨл¶„лҘҳк°Җ л°ңмғқн•ҳмҳҖлӢӨ.
- м •нҷ•лҸ„ 0.78, л№„лӢ№лҮЁлҠ” мһҳ л§һм·„кі  лӢ№лҮЁ нҒҙлһҳмҠӨлҠ” мғҒлҢҖм ҒмңјлЎң м„ұлҠҘмқҙ лӮ®м•ҳлӢӨ

```python
import pandas as pd
import numpy as np
import tensorflow as tf

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

from tensorflow.keras import layers, models
```

```python
path = '/Users/ganghyejun/Desktop/100_College/2-1/мқёкіөм§ҖлҠҘк°ңлЎ /AICLASS_2-1/task/src/data/diabetes.csv'
df = pd.read_csv(path)
df = pd.read_csv(path, index_col=0)
```

```python
y = df['Outcome']
X = df.drop('Outcome', axis=1)

# 1. XлҘј numpyлЎң ліҖнҷҳ + мҠӨмјҖмқјл§Ғ
scaler = StandardScaler()
X = scaler.fit_transform(X)

# 2. train/test л¶„лҰ¬
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=0
)

# 3. yлҘј мӣҗ-н•« мқёмҪ”л”©
classes = sorted(y.unique())
y_train = pd.get_dummies(y_train).reindex(columns=classes, fill_value=0).values
y_test = pd.get_dummies(y_test).reindex(columns=classes, fill_value=0).values

# 4. л”Ҙлҹ¬лӢқ л¶„лҘҳ лӘЁлҚё
model = models.Sequential([
    layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    layers.Dropout(0.3),
    layers.Dense(32, activation='relu'),
    layers.Dense(y_train.shape[1], activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=50,
    batch_size=16,
    verbose=1
)
```

```python
# 5. мҳҲмёЎ л°Ҹ нҸүк°Җ
y_pred = model.predict(X_test)

y_test_class = np.argmax(y_test, axis=1)
y_pred_class = np.argmax(y_pred, axis=1)

print(classification_report(y_test_class, y_pred_class))
print(confusion_matrix(y_test_class, y_pred_class))
```
