# Wine Classification

wine 데이터셋을 이용해 와인의 클래스를 분류하는 딥러닝 분류를 수행하였다.
 - 이번 테스트에서는 정확도 약 0.98로 매우 높은 성능을 보였고, 대부분의 데이터를 정확하게 분류하였다.

```python
import pandas as pd
import numpy as np
import tensorflow as tf

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

from tensorflow.keras import layers, models
```

```python
path = '/Users/ganghyejun/Desktop/100_College/2-1/인공지능개론/AICLASS_2-1/task/src/data/wine.csv'
df = pd.read_csv(path)
```

```python
y = df['Wine']
X = df.drop('Wine', axis=1)

# 1. X를 numpy로 변환 + 스케일링
scaler = StandardScaler()
X = scaler.fit_transform(X)

# 2. train/test 분리
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=0
)

# 3. y를 원-핫 인코딩
classes = sorted(y.unique())
y_train = pd.get_dummies(y_train).reindex(columns=classes, fill_value=0).values
y_test = pd.get_dummies(y_test).reindex(columns=classes, fill_value=0).values

# 4. 딥러닝 분류 모델
model = models.Sequential([
    layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    layers.Dropout(0.3),
    layers.Dense(32, activation='relu'),
    layers.Dense(y_train.shape[1], activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=50,
    batch_size=16,
    verbose=1
)
```

```python
# 5. 예측 및 평가
y_pred = model.predict(X_test)

y_test_class = np.argmax(y_test, axis=1)
y_pred_class = np.argmax(y_pred, axis=1)

print(classification_report(y_test_class, y_pred_class))
print(confusion_matrix(y_test_class, y_pred_class))
```
