# Wine Regression

wine лҚ°мқҙн„°м…Ӣмқ„ мқҙмҡ©н•ҙ м—°мҶҚнҳ• к°’мқ„ мҳҲмёЎн•ҳлҠ” нҡҢк·Җ л¬ём ңлҘј мҲҳн–үн•ҳмҳҖлӢӨ.
XлҠ” мһ…л Ҙ нҠ№м„ұмңјлЎң, yлҠ” мҳҲмёЎн•  мҲҳм№ҳ лҚ°мқҙн„°лЎң мӮ¬мҡ©н•ҳмҳҖлӢӨ.

- мқҙлІҲ л”Ҙлҹ¬лӢқ нҡҢк·Җ н…ҢмҠӨнҠёмқҳ нҸүк· м ңкіұмҳӨм°Ё(MSE)лҠ” м•Ҫ 2.95лЎң мёЎм •лҗҳм—ҲлӢӨ.

```python
import pandas as pd
import numpy as np
import tensorflow as tf

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

from tensorflow.keras import layers, models
```

```python
path = '/Users/ganghyejun/Desktop/100_College/2-1/мқёкіөм§ҖлҠҘк°ңлЎ /AICLASS_2-1/task/src/data/wine.csv'
df = pd.read_csv(path)
```

```python
y = df['Alcohol']
X = df.drop('Alcohol', axis=1)

# 1. XлҘј numpyлЎң ліҖнҷҳ + мҠӨмјҖмқјл§Ғ
scaler = StandardScaler()
X = scaler.fit_transform(X)

# 2. train/test л¶„лҰ¬
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=0
)

# 3. л”Ҙлҹ¬лӢқ нҡҢк·Җ лӘЁлҚё
model = models.Sequential([
    layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    layers.Dense(32, activation='relu'),
    layers.Dense(1)
])

model.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mse']
)

history = model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=50,
    batch_size=16,
    verbose=1
)
```

```python
# 4. мҳҲмёЎ л°Ҹ нҸүк°Җ
y_pred = model.predict(X_test)

print(mean_squared_error(y_test, y_pred))
```
