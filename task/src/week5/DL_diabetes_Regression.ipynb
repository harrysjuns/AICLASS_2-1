diabetes лҚ°мқҙн„°м…Ӣмқ„ мқҙмҡ©н•ҙ м—°мҶҚнҳ• к°’мқ„ мҳҲмёЎн•ҳлҠ” л”Ҙлҹ¬лӢқ нҡҢк·Җ л¬ём ңлҘј мҲҳн–үн•ҳмҳҖлӢӨ.
мһ…л Ҙ лҚ°мқҙн„°(X)мҷҖ м •лӢө лҚ°мқҙн„°(y)лҘј лӮҳлҲҲ л’Ө, л”Ҙлҹ¬лӢқ нҡҢк·Җ лӘЁлҚёмқ„ мӮ¬мҡ©н•ҙ к°’мқ„ мҳҲмёЎн•ҳмҳҖлӢӨ.

- мқҙлІҲ л”Ҙлҹ¬лӢқ нҡҢк·Җ н…ҢмҠӨнҠёмқҳ нҸүк· м ңкіұмҳӨм°Ё(MSE)лҠ” м•Ҫ 48.02лЎң мёЎм •лҗҳм—ҲлӢӨ.
- нҡҢк·Җ л¬ём ңмқҙлҜҖлЎң нҒҙлһҳмҠӨ л¶„лҘҳк°Җ м•„лӢҲлқј м—°мҶҚм Ғмқё мҲҳм№ҳ к°’мқ„ мҳҲмёЎн•ҳлҠ” кІғмқҙ лӘ©н‘ңмқҙлӢӨ.

```python
import pandas as pd
import numpy as np
import tensorflow as tf

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

from tensorflow.keras import layers, models
```

```python
path = '/Users/ganghyejun/Desktop/100_College/2-1/мқёкіөм§ҖлҠҘк°ңлЎ /AICLASS_2-1/task/src/data/diabetes.csv'
df = pd.read_csv(path)
```

```python
y = df['BMI']
X = df.drop(['BMI', 'Outcome'], axis=1)

# 1. XлҘј numpyлЎң ліҖнҷҳ + мҠӨмјҖмқјл§Ғ
scaler = StandardScaler()
X = scaler.fit_transform(X)

# 2. train/test л¶„лҰ¬
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0
)

# 3. л”Ҙлҹ¬лӢқ нҡҢк·Җ лӘЁлҚё
model = models.Sequential([
    layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    layers.Dense(32, activation='relu'),
    layers.Dense(1)
])

model.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mse']
)

history = model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=50,
    batch_size=16,
    verbose=1
)
```

```python
# 4. мҳҲмёЎ л°Ҹ нҸүк°Җ
y_pred = model.predict(X_test)

print(mean_squared_error(y_test, y_pred))
```
