# Wine Regression

딕셔너리 타입의 데이터셋을 라이브러리로 부터 이미 분리된 채로 포장된 문제(.data)와 정답(.target)을 각각 불러와 사용.
> 기존 타겟 레이블 대신, 데이터의 특정 컬럼 특성 하나를 새로운 정답(y)으로 설정하여 연속적인 수치 값을 예측합니다.

- 이번 회귀 테스트에선 knn 모델이 가장 낮은 수치로 성능이 높게 측정되었다.

```python
# 도구
import pandas as pd
import numpy as np

# 회귀 모델 불러오기
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor

# 회귀 모델 평가도구
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python

from sklearn import datasets
data = datasets.load_wine()
data
```

```python
y=data.target

X=data.data

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25,random_state=0)
```

```python
model_lr = LinearRegression()
model_lr.fit(X_train, y_train)

ly_preds_lr = model_lr.predict(X_test)

print ("\n--- Linear Regression Regression ---")
print('평균제곱오차', mean_squared_error(ly_preds_lr, y_test))


model_dt = DecisionTreeRegressor()
model_dt.fit(X_train, y_train)

ly_preds_dt = model_dt.predict(X_test)

print ("\n--- DecisionTree Regression ---")
print('평균제곱오차', mean_squared_error(ly_preds_dt, y_test))


model_rf = RandomForestRegressor()
model_rf.fit(X_train, y_train)

ly_preds_rf = model_rf.predict(X_test)

print ("\n--- RandomForest Regression ---")
print('평균제곱오차', mean_squared_error(ly_preds_rf, y_test))


model_kn = KNeighborsRegressor()
model_kn.fit(X_train, y_train)

ly_preds_kn = model_kn.predict(X_test)

print ("\n--- KNeighborsRegressor Regression ---")
print('평균제곱오차', mean_squared_error(ly_preds_kn, y_test))
```
